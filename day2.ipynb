## Part 1

```python
def is_increasing(l):
    l_sorted = sorted(l)
    return l == l_sorted

def is_decreasing(l):
    l_sorted = sorted(l, reverse=True)
    return l == l_sorted

def differences_ok(l):
    diffs = []
    for i in range(1, len(l)):
        diffs.append(abs(l[i]-l[i-1]))
    if max(diffs) <= 3:
        return True
    else:
        return False

def remove_duplicates(l):
    '''removes duplicate values from a sorted list l and return list'''
    return [l[i] for i in range(len(l)) if i == 0 or l[i] != l[i-1]]

def is_safe(l):
    if is_decreasing(l) or is_increasing(l):
            if differences_ok(l):
                if len(l) == len(remove_duplicates(l)):
                    return True
    else:
        return False

```

```python
n = 0
with open("day2_1_input.txt") as f:
    for line in f:
        line = line.strip()
        l = [int(x) for x in line.split(" ")]
        print(l)
        print(is_decreasing(l))
        print(is_increasing(l))
        print(differences_ok(l))
        if is_decreasing(l) or is_increasing(l):
            if differences_ok(l):
                if len(l) == len(remove_duplicates(l)):
                    n+=1

print(n) #356
```
