```python
col1 = []
col2 = []
with open("day1_1_input.csv") as f:
    for line in f:
        line = line.strip()
        elems = line.split("   ")
        col1.append(elems[0])
        col2.append(elems[1])

print(len(col1))
print(len(col2))
```

```python
print(col1[:20])
```

```python
col1.sort()
print(col1[:20])
```

```python
col2.sort()
```

```python
n = 0
for i in range(len(col1)):
    n += abs(int(col1[i])-int(col2[i]))
print(n)
```

```python
import pandas as pd
```

```python
data = {"left": col1,
        "right": col2}
df = pd.DataFrame(data, dtype=int)
```

```python
value_dict = df.right.value_counts().to_dict()
```

```python
print(value_dict)
```

```python
df["sim_score"] = df.left.apply(lambda x: x * value_dict.get(x, 0))
df["sim_score"].sum()
```
